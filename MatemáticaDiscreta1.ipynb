```python
import matplotlib.pyplot as plt

def menu_principal():
    while True:
        print("\n=== Teor铆a de conjuntos, relaciones y funciones ===")
        print("1. Caso pr谩ctico 01")
        print("2. Caso pr谩ctico 02")
        print("0. Salir")
        opcion = input("Elige una opci贸n: ")

        if opcion == '1':
            menu_caso1()
        elif opcion == '2':
            menu_caso2()
        elif opcion == '0':
            print("Saliendo... \U0001F44B")
            break
        else:
            print("Opci贸n inv谩lida. Intenta otra vez.")


def menu_caso1():
    Lista1 = [1,2,3,4,-4,-5,-6,23,25,27,100,200,240,240]
    Lista2 = [-4,-5,-6,300,400,600,500,500]

    Lista1 = list(set(Lista1))
    Lista2 = list(set(Lista2))

    while True:
        print("\n--- Caso Pr谩ctico 01 ---")
        print("1. Cardinal 1")
        print("2. Cardinal 2")
        print("3. Uni贸n")
        print("4. Diferencia 1")
        print("5. Diferencia 2")
        print("6. Intersecci贸n")
        print("0. Volver al men煤 principal")
        opcion = input("Selecciona una sub-opci贸n: ")

        if opcion == '1':
            print(f"Cardinal de Lista1: {len(Lista1)} elementos")
        elif opcion == '2':
            print(f"Cardinal de Lista2: {len(Lista2)} elementos")
        elif opcion == '3':
            union = set(Lista1).union(Lista2)
            print(f"Uni贸n: {union}")
        elif opcion == '4':
            diferencia1 = set(Lista1).difference(Lista2)
            print(f"Elementos en Lista1 pero no en Lista2: {diferencia1}")
        elif opcion == '5':
            diferencia2 = set(Lista2).difference(Lista1)
            print(f"Elementos en Lista2 pero no en Lista1: {diferencia2}")
        elif opcion == '6':
            interseccion = set(Lista1).intersection(Lista2)
            print(f"Intersecci贸n: {interseccion}")
        elif opcion == '0':
            break
        else:
            print("Opci贸n inv谩lida. Intenta de nuevo.")


def menu_caso2():
    while True:
        print("\n--- Caso Pr谩ctico 02 ---")
        print("1. Funci贸n de una sola variable")
        print("2. Funci贸n de dos variables")
        print("3. Funci贸n compuesta")
        print("4. Gr谩fica de funciones")
        print("0. Volver al men煤 principal")
        opcion = input("Selecciona una sub-opci贸n: ")

        if opcion == '1':
            funcion_una_variable()
        elif opcion == '2':
            funcion_dos_variables()
        elif opcion == '3':
            funcion_compuesta()
        elif opcion == '4':
            grafica_funcion()
        elif opcion == '0':
            break
        else:
            print("Opci贸n inv谩lida. Intenta de nuevo.")


def funcion_una_variable():
    print("\nEvaluando f(x) = x^2 + 3")
    x = int(input("Ingresa un valor para x: "))
    resultado = x**2 + 3
    print(f"f({x}) = {resultado}")


def funcion_dos_variables():
    print("\nEvaluando f(x, y) = x^2 + y^2")
    x = int(input("Ingresa valor para x: "))
    y = int(input("Ingresa valor para y: "))
    resultado = x**2 + y**2
    print(f"f({x},{y}) = {resultado}")


def funcion_compuesta():
    print("\nFunciones f(x) = 2x+1 y g(x) = x^2")
    x = int(input("Ingresa un valor para x: "))
    gx = x**2
    fx = 2*gx + 1
    print(f"g({x}) = {gx}")
    print(f"f(g({x})) = {fx}")


def grafica_funcion():
    print("\nGraficando f(x) = x^2 + 2")
    x = list(range(-10,11))
    y = [i**2 + 2 for i in x]

    plt.plot(x, y, marker='o')
    plt.title('Gr谩fica de f(x) = x^2 + 2')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.show()


if __name__ == "__main__":
    menu_principal()
```
